```python
import pandas as pd
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
```

```python
my_url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_Hindi_films"
```

```python
my_web_page = urlopen(my_url).read()
```

```python
page_soup = soup(my_web_page)
```

```python
page_soup
```

```python
table = page_soup.find(name = 'table', attrs = {'class' : 'wikitable'})
table
```

```python
wiki_data = []

rows = table.find_all('tr')
```

```python
for row in rows:
    cells = row.find_all(['td', 'th'])
    row_data = []
    for cell in cells:
        row_data.append(cell.get_text(strip=True))
    wiki_data.append(row_data)
```

```python
wiki_data
```

```python
df = pd.DataFrame(wiki_data)
df.head()
```

```python
# CONVERT TO CSV FILE

df.to_csv('highest_grossing_hindi_movies', index = False)
```
